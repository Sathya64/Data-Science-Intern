```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset from CSV
df = pd.read_csv("C:\\Users\\welcome\\Downloads\\Desktop\\task 2.zip")

# Create a bar chart using Seaborn
sns.barplot(x='training_samples', y='test_samples', data=df, palette='viridis')

# Adding labels and title
plt.xlabel('training_samples')
plt.ylabel('test_samples')
plt.title('Bar Chart Example with Seaborn')

# Display the chart
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset from CSV
df = pd.read_csv("C:\\Users\\welcome\\Downloads\\Desktop\\task 2.1.zip")

# Create a line chart using Seaborn
sns.lineplot(x='Age', y='Job', data=df, marker='o', color='blue')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Job')
plt.title('Line Chart Example with Seaborn')

# Display the chart
plt.show()
```
